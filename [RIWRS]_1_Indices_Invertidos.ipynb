```python
import pandas as pd
import string
artigos = pd.read_json('https://raw.githubusercontent.com/lucianomcsilva/articulistas/main/alexandre_garcia.json')
# remocação de pontuaçao
artigos['title'] = artigos['title'].str.lower()
artigos['title'] = artigos['title'].str.replace('[{}]'.format(string.punctuation), '')
artigos['title'] = artigos['title'].str.replace('\n', ' ')
artigos['title'] = artigos['title'].str.replace('“', ' ')
artigos['title'] = artigos['title'].str.replace('”', ' ')
artigos['title'] = artigos['title'].str.replace("'", ' ')
artigos['title'] = artigos['title'].str.replace("‘", ' ')
artigos['title'] = artigos['title'].str.replace("’", ' ')
artigos['title'] = artigos['title'].str.replace("’", ' ')
artigos['title'] = artigos['title'].str.replace("alexandre garcia", ' ')
#artigos['title'] = artigos['title'].str.lower().apply(lambda x: ' '.join([word for word in x.split() if word not  in (string.ascii_letters)]))
artigos['title'] = artigos['title'].str.lower().apply(lambda x: x.strip())


display(pd.DataFrame(artigos['title']).head(50))
titles = list(artigos['title'])
artigos.shape
```

```python
inverted_index = {}
for index, title in enumerate(titles):  
  for word in title.split(' '):
    if word not in inverted_index:
      inverted_index.update({word: {"document": [index]} })    
    else:
      inverted_index[word]["document"].append(index)


# print(inverted_index)
# import pprint
# pp = pprint.PrettyPrinter(depth=4)
# pp.pprint(inverted_index)

print(inverted_index['supremo'])
    
```
